```python
# Import required modules
import pandas as pd
import numpy as np
from statsmodels.formula.api import logit

# Start coding!
car_insurance = pd.read_csv("car_insurance.csv")
display(car_insurance.head())
print(car_insurance.info())
```

```python
# handling missing data
car_insurance.credit_score.fillna(car_insurance.credit_score.mean(),inplace=True)
car_insurance.annual_mileage.fillna(car_insurance.annual_mileage.mean(),inplace=True)
print(car_insurance.info())
```

```python
def model_accuracy(df):
    """
        this function builds a logistic regression model and calculates the accuracy
        
        Return the Accuracy
        
        NOTES :
            the first column has to be the explanatory variable
            the sacond column has to be the response variable
    """
    from statsmodels.formula.api import logit
    formula = df.columns[1] + " ~ " + df.columns[0]
    model = logit(formula=formula, data=df).fit()
    
    conf_matrix = model.pred_table()
    TN = conf_matrix[0,0]
    TP = conf_matrix[1,1]
    FN = conf_matrix[1,0]
    FP = conf_matrix[0,1]
    
    acc = (TN + TP) / (TN + TP + FN + FP)
    return acc
```

```python
accuracy = {}
for i in car_insurance.columns[1:-1]:
    accuracy.update({i:model_accuracy(car_insurance[[i,"outcome"]])})
    
print(accuracy)
```

```python

df = pd.DataFrame.from_dict(accuracy,orient='index', columns=["accuracy"])\
.reset_index().sort_values(by="accuracy", ascending=False)

print(df)
```
